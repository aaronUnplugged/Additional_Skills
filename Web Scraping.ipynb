```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
from pandas import Series,DataFrame
```

```python
url = 'http://www.bbc.com/sport/football/tables'
```

```python
result = requests.get(url)
c = result.content
```

```python
soup = BeautifulSoup(c,"lxml")
```

```python
summary = soup.find('div',{'class':'stats', 'id':'blq-content'})
tables = summary.find_all('table')
```

```python
data = []

rows = tables[0].findAll('tr')

for tr in rows:
    cols = tr.findAll('td')
    
    for td in cols: 
        
        text = td.find(text=True)
        print text,
        data.append(text)
```

```python
#Data scraped from the web
data
```

```python
#Pulling the headers from the web table
headers = [th.getText() for th in
          soup.findAll('tr',limit=2)[0].findAll('th')]
```

```python
headers
```

```python
data_rows = soup.findAll('tr')[1:]
player_data = [[td.getText() for td in data_rows[i].findAll('td')]
              for i in range(len(data_rows))]
```

```python
#Recreating web table as a panda data frame
df = pd.DataFrame(player_data, columns = headers)
```

```python
df.head()
```

```python
#Cleaning up the table
df.drop(['Position','Last 10 games results','Match status'],axis=1,inplace=True)
df.drop([0],inplace=True)
```
